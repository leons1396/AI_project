```python
# Bibliotheken importieren

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# Bild importieren

img_bgr = cv.imread(r'C:\Users\peter\Downloads\Testbilder\kartoffel_trieb_185_cut_red.jpg')
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# in HSV wandeln

img_hsv = cv.cvtColor(img_rgb, cv.COLOR_RGB2HSV)


# Triebe einfärben
# Vorschlag 2

# H = (len(img_hsv))         # entspricht Höhe
# B = (len(img_hsv[0]))      # entspricht Breite

for i in range(len(img_hsv)):
    for j in range(len(img_hsv[0])):
        Farben = img_hsv[i][j]
        if Farben[0] > 19 and Farben[0] < 25 and Farben[1] > 100 and Farben[1] < 130:         
            Farben[0] = 41
            Farben[1] = 35
            Farben[2] = 100
            img_hsv[i][j] = Farben
        else:
            img_hsv[i][j] = (0,0,0)
            
img_rgb2 = cv.cvtColor(img_hsv, cv.COLOR_HSV2RGB)
plt.imshow(img_rgb2)
```
