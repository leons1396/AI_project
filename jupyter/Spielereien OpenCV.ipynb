```python
# Bibliotheken importieren

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# Bild importieren

img_bgr = cv.imread(r'C:\Users\peter\Downloads\Testbilder\kartoffel_trieb_89_cut.jpeg')
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB) # somit wird gleich richtig eingelesen, openCV arbeitet sonst mit BGR
plt.imshow(img_rgb)
```

```python
# Gaussian Blur

img_blr = img_rgb.copy()

img_blr_2 = cv.GaussianBlur(img_blr, (291,291),0)

plt.imshow(img_blr_2)
```

```python
# Alle Pixel bei denen sowohl Rot, als auch Grn und Blau grer ca. 175 sind WEISS einfrben
for i in range(len(img_blr_2)):
    for j in range(len(img_blr_2[0])):
        Farben = img_blr_2[i][j]
        #print(Farben)
        if Farben[0] >= 130 and Farben[1] >= 130 and Farben[2] >= 130:      
            # Wert muss noch verfeinert werden
            # bei meiner Beleuchtung wrde ich eine hhere Schwelle fr Rot nehmen => 190 / 175 / 175
            # Schwellenwert ist stark von Beleuchtung abhngig ... jedesmal Schwelle anpassen aber dumm ... 2-4 Klassen  la Leon drfte sinniger sein
            img_blr_2[i][j] = (255,255,255)

plt.imshow(img_blr_2)
```

```python
# Median Blur

img_blr = img_rgb.copy()

img_blr_3 = cv.medianBlur(img_blr, 37)

plt.imshow(img_blr_3)
```

```python
# Alle Pixel bei denen sowohl Rot, als auch Grn und Blau grer ca. 175 sind WEISS einfrben
for i in range(len(img_blr_3)):
    for j in range(len(img_blr_3[0])):
        Farben = img_blr_3[i][j]
        #print(Farben)
        if Farben[0] >= 130 and Farben[1] >= 130 and Farben[2] >= 130:      
            # Wert muss noch verfeinert werden
            # bei meiner Beleuchtung wrde ich eine hhere Schwelle fr Rot nehmen => 190 / 175 / 175
            # Schwellenwert ist stark von Beleuchtung abhngig ... jedesmal Schwelle anpassen aber dumm ... 2-4 Klassen  la Leon drfte sinniger sein
            img_blr_3[i][j] = (255,255,255)

plt.imshow(img_blr_3)
```

```python
# Sobel

img_sbl = img_rgb.copy()

img_sbl = cv.Sobel(img_sbl, -1, 0,1)

plt.imshow(img_sbl)
```

```python
# Scharr

img_schrr = img_rgb.copy()

img_schrr = cv.Scharr(img_schrr,-1,1,0)

plt.imshow(img_schrr)
```

```python
img_schrr_2 = cv.medianBlur(img_schrr, 5)

plt.imshow(img_schrr_2)
```

```python
# Alle Pixel bei denen sowohl Rot, als auch Grn und Blau grer ca. 175 sind WEISS einfrben
for i in range(len(img_schrr_2)):
    for j in range(len(img_schrr_2[0])):
        Farben = img_schrr_2[i][j]
        #print(Farben)
        if Farben[0] < 5 and Farben[1] < 5 and Farben[2] < 5:      
            # Wert muss noch verfeinert werden
            # bei meiner Beleuchtung wrde ich eine hhere Schwelle fr Rot nehmen => 190 / 175 / 175
            # Schwellenwert ist stark von Beleuchtung abhngig ... jedesmal Schwelle anpassen aber dumm ... 2-4 Klassen  la Leon drfte sinniger sein
            img_schrr_2[i][j] = (0,255,0)

plt.imshow(img_schrr_2)
```
