```python
# Bibliotheken importieren

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# Bild importieren

img_bgr = cv.imread(r'C:\Users\peter\Downloads\Testbilder\kartoffel_trieb_163_cut.jpeg')
gray = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY) # somit wird gleich richtig eingelesen, openCV arbeitet sonst mit BGR
plt.imshow(gray)
```

```python
# SURF-Objekt erstellen
sift = cv.SIFT_create(50000)  #je höher der Wert, desto mehr Sachen findet er

# Keypoints und Deskriptoren mit SIFT extrahieren
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Bild mit Keypoints zeichnen
img_w_kp = cv.drawKeypoints(gray,keypoints,None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

#Ergebnis anzeigen
plt.imshow(img_w_kp)

print(f"Anzahl der Keypoints: {len(keypoints)}")
```

```python
# Ausgabe der Keypoints
count = 0
criteria = 10

for i in range(len(keypoints)):
       act_keypoint = keypoints[i]
       size = act_keypoint.size
       if size >= criteria:
              x,y = act_keypoint.pt
              print(f"Position des {i}. Keypoints: x = {x}, y = {y}")
              print(f"Groesse des {i}. Keypoints = {size}")
              angle = act_keypoint.angle
              print(f"Winkel des {i}. Keypoints = {angle}")
              count = count+1

print(f"Anzahl der Keypoints groesser {criteria} beträgt {count}")       
```
