```python
img_green_2 = img_rgb.copy()
H = (len(img_green_2))         # entspricht Hhe
B = (len(img_green_2[0]))      # entspricht Breite
count = 0

for i in range(len(img_green_2)):
    for j in range(len(img_green_2[0])):
        Farben = img_green_2[i][j]
        if Farben[0] > 170 and Farben[1] > 165 and Farben[2] <150:                 
            # anderer Ansatz: Grn ist es, wenn mehr Grn als andere Farben enthalten ist
            Farben[0] = 0
            Farben[1] = 255
            Farben[2] = 0
            img_green_2[i][j] = Farben
        else:
            img_green_2[i][j] = (0,0,0)
            
plt.imshow(img_green_2)

print(count)
print(count/(H*B)*100)
```
