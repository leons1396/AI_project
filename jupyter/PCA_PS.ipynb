```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
training_df = pd.read_csv(r'C:\Users\peter\Downloads\vegis_traindata.csv')
training_df=training_df.drop(columns=["Unnamed: 0","Image"])

# Neues Feature = Aspect Ration
aspectratio_df = training_df.loc[:]['Hight']/training_df.loc[:]['Width']
training_df['Aspect ratio'] = aspectratio_df
col = training_df.pop('Aspect ratio')
training_df.insert(loc=2, column='Aspect ratio', value=col)
training_df.head()
```

```python
x_train = training_df.drop(columns=["Label"])
print ('x_train: ', x_train.shape)
```

```python
x_train.head()
```

```python
# Skalierung
x_train_scaled = (x_train - x_train.mean()) / x_train.std()
x_train_scaled.head()
```

```python
# Eigentliche PCA
pca = PCA(n_components=10)
pca.fit(x_train_scaled)
```

```python
print(f"Rotation matrix:\n{pca.components_}")
```

```python
PCA_df = pd.DataFrame({'1. HK': pca.components_[0,:], 
                       '2. HK': pca.components_[1,:], 
                       '3. HK': pca.components_[2,:],
                       '4. HK': pca.components_[3,:],
                       '5. HK': pca.components_[4,:],
                       '6. HK': pca.components_[5,:],
                       '7. HK': pca.components_[6,:],
                       '8. HK': pca.components_[7,:],
                       '9. HK': pca.components_[8,:],
                       '10. HK': pca.components_[9,:]})
PCA_df.insert(loc=0, column='Feature', value = ['Hight','Width','Aspect ratio','R','G','B','green pixels','green onion pixels','ratio area','circularity','keypoints','mean keypoints'])
PCA_df.head(12)
```

```python
#PCA_df.to_excel(r'C:\Users\peter\Downloads\PCA.xlsx')
```

```python
PCA_2_df = pd.DataFrame({'Total': pca.explained_variance_[:]})
PCA_2_df.insert(loc=0, column='Komponente', value = ['1.HK','2.HK','3.HK','4.HK','5.HK','6.HK','7.HK','8.HK','9.HK','10.HK'])
letzte_Spalte_df = pd.DataFrame({'Prozent der Varianz': pca.explained_variance_ratio_[:]})
PCA_2_df['Prozent der Varianz'] = letzte_Spalte_df
PCA_2_df.head(12)
```
