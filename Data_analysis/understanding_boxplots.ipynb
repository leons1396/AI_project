```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

# https://builtin.com/data-science/boxplot

```python
arr = np.arange(1, 12, 1)
print(arr)
arr = np.append(arr, 20)
for _ in range(1, 8):
    x = 2
    arr = np.append(arr, x)

for _ in range(1, 2):
    x = 10
    arr = np.append(arr, x)
print("Mean: ", arr.mean())
print(arr)
```

```python
np.median(arr)
```

```python
sns.boxplot(data=arr)
```

```python
arr = np.arange(1, 12, 1)
print(arr)
arr = np.append(arr, 20)
sns.boxplot(data=arr)
highlighted_point_y = np.arange(10, 12, 1)
highlighted_point_x = np.array([0, 1])
plt.scatter(highlighted_point_x, highlighted_point_y, color='red', marker='o', label='Highlighted Point')
plt.xlabel('Boxplot')
plt.ylabel('Values')
plt.title('Boxplot with Highlighted Data Point')

# Showing the plot
plt.legend()
plt.show()
```
